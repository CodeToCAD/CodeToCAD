```python
# Testing build123d
from build123d import Box, Rectangle, extrude

length, width, thickness = 1, 1, 1

# cube = Box(length, width, thickness)
my_sketch = Rectangle(width=1, height=1)
cube = extrude(my_sketch, amount=1)
```

```python
from platform import system

from build123d import export_stl
import open3d as o3d

if system() != "Darwin":
    from open3d.web_visualizer import draw
else:
    # The web visualizer is not available on macOS.
    from open3d.visualization import draw


filepath = "./cube.stl"
export_stl(cube, filepath)

mesh = o3d.io.read_triangle_mesh(filepath)
mesh.paint_uniform_color([0.5, 0.5, 0.5])  # Gray color

draw(mesh)
```

```python
cube = Part.preset.cube(1, 1, 1)

print(f"{cube=}")

from build123d import BuildLine, BuildPart, BuildSketch, Line, Spline, Vector, make_face

from codetocad.cad.wire.wire_operations import WireOperationLineTo


with BuildPart() as ex12:
    with BuildSketch() as ex12_sk:
        for wire in cube.sketch.wires:
            vertices = wire.get.vertices
            edges = wire.get.edges

            for operation in wire._operations:
                if isinstance(operation, WireOperationLineTo):
                    with BuildLine() as ex12_ln:
                        last = ex12_ln.line.vertices() if ex12_ln.line else (0.0, 0.0)
                        l2 = Line(
                            last,
                            (operation.x.value, operation.y.value),
                        )

                make_face()

                extrude(amount=10)
```
