# CodeToCAD

## Automate Manufacturing — Without Replacing Your Tools

CodeToCAD started in a garage while I was working with a friend on converting a gas go-kart into an electric autonomous vehicle.

I needed to model the chassis to mount motors and a chain drive. After hours sketching, sweeping, and managing messy geometry — I had a crude replica. As a programmer, I thought: this could have been done in a few lines of code.

The result of that is CodeToCAD: Simple, repeatable, version-controlled modeling, animation, simulation and deployment. Digital Manufacturing turbo-powered by code.

Over time, a few people joined to contribute. Some were paid, others volunteering to an open-source project.

Thank you for being interested in CodeToCAD!

```python
!uv venv # Create a virtual environment using uv

# On Mac: brew install mupdf zlib llvm@16; and export their brew paths before installing pybullet: https://github.com/ahrm/sioyek/issues/1361#issue-2974068169
if __import__("platform").system() == "Darwin" and input(
    "OK to run brew install mupdf zlib llvm@16? Not installing them might cause pybullet to fail to install. (y/n): "
).strip().lower() in ("y", "yes"):
    import os

    !brew install mupdf zlib llvm@16
    # Update PATH
    os.environ["PATH"] = f"/opt/homebrew/opt/llvm@16/bin:{os.environ.get('PATH', '')}"

    # Set compiler paths
    os.environ["CC"] = "/opt/homebrew/opt/llvm@16/bin/clang"
    os.environ["CXX"] = os.environ["CC"] + "++"

    # Update flags
    os.environ["LDFLAGS"] = (
        f"{os.environ.get('LDFLAGS', '')} -L/opt/homebrew/opt/llvm@16/lib"
    )
    os.environ["CPPFLAGS"] = (
        f"{os.environ.get('CPPFLAGS', '')} -I/opt/homebrew/opt/llvm@16/include"
    )

!uv sync --all-groups --all-extras # Install dependencies
```
