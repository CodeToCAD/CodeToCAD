# CodeToCAD

## Automate Manufacturing — Without Replacing Your Tools

CodeToCAD started in a garage while I was converting a gas go-kart into an electric autonomous vehicle.

I needed to model the chassis to mount motors and a chain drive. After hours sketching, sweeping, and managing messy geometry — I had a crude replica. As a programmer, I thought: this could have been done in a few lines of code.

The result of that is CodeToCAD: Simple, repeatable, version-controlled modeling, animation, simulation and deployment. Digital Manufacturing turbo-powered by code.

Overtime, a few people joined to contribute. Some were paid, others volunteering to an open-source project.

```python
"""
Installs a pre-commit hook to run auto-formatting and test before committing code.

See https://pre-commit.com/ and https://git-scm.com/book/en/v2/Customizing-Git-Git-Hooks for more info.
"""

import subprocess

root_dir = subprocess.run(
    "git rev-parse --show-toplevel".split(" "), capture_output=True, text=True
).stdout.strip()

pre_commit_filepath = f"{root_dir}/.git/hooks/pre-commit"

if __name__ == "__main__":
    """
    This script sets up a pre-commit hook to run auto-formatting and test.
    """
    with open(pre_commit_filepath, "w") as f:
        f.write(
            """#!/bin/bash  
set -e
                
ROOT_DIR=`git rev-parse --show-toplevel`

cd "$ROOT_DIR"
                
uv run black . --check
uv run pytest .
"""
        )

    # Make the pre-commit file executable
    subprocess.run(["chmod", "+x", pre_commit_filepath])
```
